Import libraries

```python
from pathlib import Path

import pandas as pd
import numpy as np
import requests

import pandas_profiling
```

```python
if __name__ == "__main__":
    file_name = Path("census.csv")
    if not file_name.exists():
        data = requests.get(
            "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
        )
        file_name.write_bytes(data.content)
```

Load data

```python
    # Names based on https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.names
    # age	workclass	education_level	education-num	marital-status	occupation	relationship	
    # race	sex	capital-gain	capital-loss	hours-per-week	native-country	income
    df = pd.read_csv(
        file_name,
        header=None,
        index_col=False,
        names=[
            "age",
            "workclass",
            "education_level",
            "education-num",
            "marital-status",
            "occupation",
            "relationship",
            "race",
            "sex",
            "capital-gain",
            "capital-loss",
            "hours-per-week",
            "native-country",
            "income"
        ],
    )
```

Prepare missing values

```python
df = df.replace("\\?", np.nan, regex=True)
```

Run Pandas Profiler

```python
 profile = df.profile_report(title="Census Dataset")
```
